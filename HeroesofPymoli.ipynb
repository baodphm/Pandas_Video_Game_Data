```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

purchase_data_df.head(25)
```

## Player Count

* Display the total number of players


```python
Unique_Players_df = purchase_data_df.groupby('SN').first().reset_index()

Total_Players = len(Unique_Players_df)
Dict_Players = {'Total Players': [Total_Players]}
Players_df = pd.DataFrame.from_dict(Dict_Players)
Players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_Items = purchase_data_df["Item ID"].nunique()
Average_Price = purchase_data_df["Price"].mean()
Number_Purchase = len(purchase_data_df)
Total_Revenue = purchase_data_df["Price"].sum()
```

```python
Dict_Purchasing = {
    'Number of Unique Items': [Unique_Items],
    'Average Price': [Average_Price],
    'Number of Purchases': [Number_Purchase],
    'Total Revenue': [Total_Revenue]}


Analysis_df = pd.DataFrame.from_dict(Dict_Purchasing)
pd.options.display.float_format = '${:.2f}'.format
Analysis_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Bros = Unique_Players_df.loc[Unique_Players_df['Gender']== 'Male', 'Price'].count()
Gals = Unique_Players_df.loc[Unique_Players_df['Gender']== 'Female', 'Price'].count()
Non = Unique_Players_df.loc[Unique_Players_df['Gender']== 'Other / Non-Disclosed', 'Price'].count()

Percent_Bros = Bros/Total_Players
Percent_Gals = Gals/Total_Players
Percent_Non = Non/Total_Players

gender_data = {'Gender': ["Male", "Female", "Other/Non-Disclosed"],
                'Total Count': [Bros, Gals, Non],
                'Percentage of Players': [Percent_Bros, Percent_Gals, Percent_Non]}
gender_df = pd.DataFrame(gender_data,columns = ['Gender','Total Count','Percentage of Players'])
gender_df['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val*100)for val in gender_df['Percentage of Players']], index = gender_df.index)

gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
guys = purchase_data_df.loc[purchase_data_df['Gender']== 'Male', 'Price'].count()
dolls = purchase_data_df.loc[purchase_data_df['Gender']== 'Female', 'Price'].count()
other = purchase_data_df.loc[purchase_data_df['Gender']== 'Other / Non-Disclosed', 'Price'].count()

guys_sum = purchase_data_df.loc[purchase_data_df['Gender']== 'Male', 'Price'].sum()
dolls_sum = purchase_data_df.loc[purchase_data_df['Gender']== 'Female', 'Price'].sum()
other_sum = purchase_data_df.loc[purchase_data_df['Gender']== 'Other / Non-Disclosed', 'Price'].sum()

guys_avg_purchase = guys_sum/ guys
dolls_avg_purchase = dolls_sum/ dolls
other_avg_purchase = other_sum/ other

guys_Avg_total_Purchase = guys_sum/ Bros
dolls_Avg_total_Purchase = dolls_sum/ Gals
other_Avg_total_Purchase = other_sum/ Non

gender_Purchase_data = {'Gender': ["Male", "Female", "Other/Non-Disclosed"],
                        'Purchase Count': [dolls, guys, other],
                        'Average Purchase Price': [dolls_avg_purchase, guys_avg_purchase, other_avg_purchase],
                        'Total Purchase Value': [dolls_sum, guys_sum, other_sum],
                        'Avg Total Purchase per Person': [dolls_Avg_total_Purchase, guys_Avg_total_Purchase, other_Avg_total_Purchase]}
gender_Purchase_df = pd.DataFrame(gender_Purchase_data,columns = ['Gender','Purchase Count','Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person'])

pd.options.display.float_format = '${:.2f}'.format
gender_Purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [-np.inf, 9, 14, 19, 24, 29, 34, 39, np.inf]
group_Labels = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
pd.cut(Unique_Players_df["Age"], bins, labels=group_Labels).head()

Unique_Players_df["View Group"] = pd.cut(Unique_Players_df["Age"], bins, labels=group_Labels)
Age_Group_df = Unique_Players_df.groupby("View Group")

Age_Total_Count = Age_Group_df["Age"].count()
Age_Percent = (Age_Total_Count / Total_Players) * 100

pd.options.display.float_format = '{:.2f}%'.format

Demographics_df =pd.DataFrame({
    "Total Count": Age_Total_Count,
    "Percentage of Players": Age_Percent})
Demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_df["Purchasing Analysis"] = pd.cut(purchase_data_df["Age"], bins, labels=group_Labels)
Purchasing_df = purchase_data_df.groupby("Purchasing Analysis")

Purchase_Count = Purchasing_df["Purchase ID"].count()
Avg_Purchase_Price = Purchasing_df["Price"].mean()
Total_Purchase_Value = Purchasing_df["Price"].sum()
Avg_Total_Purchase = Total_Purchase_Value/ Age_Total_Count

Age_Purchasing_df =pd.DataFrame({
                                "Purchase Count": Purchase_Count,
                                "Average Purchase Price": Avg_Purchase_Price,
                                "Total Purchase Value": Total_Purchase_Value,
                                "Average Total Purchase per Person": Avg_Total_Purchase})

pd.options.display.float_format = '${:.2f}'.format
Age_Purchasing_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Top_Spenders = purchase_data_df.groupby("SN")

Spender_Count = Top_Spenders["SN"].count()
Spender_Total = Top_Spenders["Price"].sum()
Spender_Purchase = Spender_Total / Spender_Count

Top_Spenders_df =pd.DataFrame({
                                "Purchase Count": Spender_Count,
                                "Average Purchase Price": Spender_Purchase,
                                "Total Purchase Value": Spender_Total})

Top_Spenders_df.sort_values(by=['Total Purchase Value'], inplace=True, ascending=False)
Top_Spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Popular_Items = purchase_data_df.groupby(['Item ID', 'Item Name'])

Pop_Count = Popular_Items["Item ID"].count()
Pop_Price = Popular_Items["Price"].mean()
Pop_Total = Pop_Count * Pop_Price

Most_Pop_df =pd.DataFrame({
                            "Purchase Count": Pop_Count,
                            "Item Price": Pop_Price,
                            "Total Purchase Value": Pop_Total})

Most_Pop_df.sort_values(by=['Purchase Count'], inplace=True, ascending=False)
Most_Pop_df.head()
```
